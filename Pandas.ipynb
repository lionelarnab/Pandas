```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df
```

```python
df.dropna() # if we pass "how = all" then if all row/col contain NaN/NaT, then only dropping will be done.
```

```python
df
```

```python
df1
```

```python
df1.drop_duplicates()
```
